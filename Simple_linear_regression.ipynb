```python
import statsmodels.api as sm
```

```python
X = df[['TV']]
X[0:5]
```

```python
X = sm.add_constant(X)
```

```python
X[0:5]
```

```python
y = df['sales']
```

```python
y[0:5]
```

```python
lm = sm.OLS(y, X)
```

```python
model = lm.fit()
```

```python
model.summary()
```

```python
import statsmodels.formula.api as smf
lm = smf.ols('sales ~ TV', df)
```

```python
model = lm.fit()
model.summary()
```

```python
model.params
```

```python
model.summary().tables[1]
```

```python
model.conf_int()
```

```python
model.f_pvalue
```

```python
print('fvalue: ', "%.2f" % model.fvalue)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['TV']]
y = df['sales']
```

```python
reg = LinearRegression()
```

```python
model= reg.fit(X, y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(X, y)
```

```python
model.predict([[30]])
```

```python
new_data = [[5], [90], [200]]

model.predict(new_data)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lm = smf.ols('sales ~ TV', df)
model = lm.fit()
```

```python
mse = mean_squared_error(y, model.fittedvalues)  #texmin edilen deyerlerin texmini xetasi
mse
```

```python
import numpy as np

rmse = np.sqrt(mse)
rmse
```
